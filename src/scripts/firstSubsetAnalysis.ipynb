Going back to the loans that have not charged-off, let's test (visually) if it does follow a uniform distribution. 

```python
# function to get cdf of the variable
def evalCdf(ts, x):
    count = 0
    for t in ts:
        if x > t:
            count += 1
    return count /  len(ts)
```

```python
# cdf of the data
data = loans["funded_amnt"][loans_sub.loan_status == "Charged Off"]
data_cdf = sorted([evalCdf(ts = data, x = i) for i in data])
```

```python
# model data and cdf
model_data = np.random.uniform(low = data.min(), high = data.max(), size = data.shape[0])
model_cdf = sorted([evalCdf(ts = model_data, x = i) for i in model_data])
```

```python
plt.plot(sorted(data), data_cdf, color = "steelblue", label = "Data")
plt.plot(sorted(model_data), model_cdf, color = "green", label = "Model")
plt.legend(loc = "lower right")
plt.show()
```

The data we have do deviate significantly from a uniform distribution with the same parameters. The deviation is significant right at the middle of the distribution. This goes to show that some loan amounts might have higher / lower default rates than others.

Next up, I'll add in the interest rates into the mix and see what happens over there.

```python
loans_sub["int_rate"] = loans["int_rate"]
```

There seem to be some issues with interst rate. I'll remove the `%` symbol, convert it to float32 (to save space) and continue the analysis.

```python
# converting loans_sub.int_rate to float16
loans_sub["int_rate"] = loans_sub["int_rate"].str.replace("%", "")
loans_sub["int_rate"] = loans_sub["int_rate"].astype("float32")
```

```python
plt.hist(loans_sub["int_rate"][loans_sub.loan_status == "Charged Off"], label = "Charged Off", 
         color = "steelblue", alpha = 0.5)
plt.hist(loans_sub["int_rate"][loans_sub.loan_status != "Charged Off"], label = "Not Charged Off", 
         color = "green", alpha = 0.3)
plt.legend()
plt.show()
```

Hmmm... This time, the interest rates for the charged off loans seem to follow a normal distribution. 

```python
data = loans_sub["int_rate"][loans_sub.loan_status == "Charged Off"]
data_cdf = [evalCdf(ts = data, x = i) for i in data]

model = np.random.normal(loc = data.mean(), scale = data.std(), size = data.shape[0])
model_cdf = [evalCdf(ts = model, x = i) for i in model]
```

```python
plt.plot(sorted(data), sorted(data_cdf), color = 'steelblue', label = "Data")
plt.plot(sorted(model), sorted(model_cdf), color = "green", label = "Model")
plt.legend(loc = "lower right")
plt.show()
```

This data seems to agree a lot with the normal distribution. However, it does deviate quite significantly at the lower end of the distribution. Let's start off with a test in that case.

```python
# a function to carry out a permutation test
def meanDiff(x, n1, n2):
    n = n1 + n2
    total = np.arange(n)
    indices1 = np.random.choice(n, n1, replace = False)
    indices2 = total[~np.isin(total, indices1)]
    
    mean_diff = x[indices1].mean() - x[indices2].mean()
    return mean_diff

def permutationTest(x, n1, n2, times= 1000):
    differences = []
    print("Carrying out the permutation test {0} times...".format(times))
    for i in list(range(times)):
        difference = meanDiff(x, n1, n2)
        differences.append(difference)
    
    return differences
```

```python
interest_test_results = permutationTest(x = loans_sub["int_rate"], n1 = 5627, n2 = 34090)
```

```python
actual_difference = loans_sub["int_rate"][loans_sub.loan_status == "Charged Off"].mean() - loans_sub["int_rate"][loans_sub.loan_status != "Charged Off"].mean()
```

```python
plt.hist(interest_test_results, alpha = 0.5, color = "steelblue")
plt.vlines(x = actual_difference, ymin = 0, ymax = 250)
plt.show()
```

```python
# the p-value
1 - (sum(actual_difference > interest_test_results) / 1000) 
```

That's a p-value of 0.0. So, the difference in interest rates between loans that have been charged off and those that have been not charged off are statistically significant.
